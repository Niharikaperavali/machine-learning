```python
import pandas as pd
from scipy import stats
data={
    'marks':[98,78,67,57,46,34]
}
df=pd.DataFrame(data)
print(df)
mean_mark=df['marks'].mean()
print(" the mean value:",mean_mark)
median_marks=df['marks'].median()
print(" the median value:",median_marks)
mode_marks=df['marks'].mode()[0]
print("the mode value:",mode_marks)
```

```python
import pandas as pd
from scipy.stats import skew,kurtosis
data={
    'marks':[3,6,8,3,7,9,7]
}
df=pd.DataFrame(data)
print(df)
variance=df['marks'].var()
print("the variance of the marks:",variance)
standard=df['marks'].std()
print("the standard deviation of the marks:",standard)
skewness=skew(df['marks'])
print("the skewness of the marks:",skewness)
kurtosis=kurtosis(df['marks'])
print("the kurtosis of the marks:",kurtosis)
```

```python
import numpy as np
import pandas as pd
data={
    'name':['alice','bob',None,'chrlie','neha'],
    'age':[8,7,6,5,np.nan],
    'email':['alice@gmail.com','bob@gmail.com','chrlie@gmail.com',None,'neha@gmail.com'],
    'salary':[90000,89000,78000,'not available',67000]

}
df=pd.DataFrame(data)
print(df)
df['name'].fillna('unknown')
df['age'].fillna(df['age'].mean())
df['email']=df['email'].replace('unknown',np.nan)
df.drop_duplicates()
df['salary']=pd.to_numeric(df['salary'],errors='coerce')
df['salary']=df['salary'].fillna(df['salary'].mean())
df.columns=df.columns.str.strip().str.lower()
print("\ncleaned dataset:\n",df)
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.weightstats import ztest,ttest_ind
np.random.seed(0)
g1=np.random.normal(loc=50,scale=5,size=30)
g2=np.random.normal(loc=52,scale=5,size=30)
z_test,z_p=ztest(g1,g2)
print("ztest statistics")
print(f"z-statistics:{z_test:.2f},p-value:{z_p:.4f}")
t_test,t_p,t_df=ttest_ind(g1,g2)
print("ttest statistics")
print(f"ttststatic:{t_test:.2f},p-value:{t_p:.4f},df-value:{t_df:.4f}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
data={
    'experience':[7,8,4,5,6],
    'salary':[90000,769000,70000,6700000,600000]
}
df=pd.DataFrame(data)
x,y=df[['experience']],df['salary']
model=LinearRegression().fit(x,y)
y_pred=model.predict(x)
print("r2_score",r2_score(y,y_pred))
print("mae",mean_absolute_error(y,y_pred))
print("mse",mean_squared_error(y,y_pred))
```

```python
import pandas as pd
df1 = pd.DataFrame({
    'student_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David']
})

df2 = pd.DataFrame({
    'student_id': [1, 2, 3, 4],
    'math_score': [90, 80, 85, 70],
    'science_score': [88, 75, 95, 65]
})
merge_df=pd.merge(df1,df2,on='student_id')
print("merged data",merge_df)
reshape_df=merge_df.set_index(['student_id','name']).stack()
print("reshaping of the data",reshape_df)
dup_df=pd.DataFrame({
    'student_id':[9,8,9,9,6],
    'name':['ali','monni','bob','ali','neha']
})
dupe_df=dup_df.drop_duplicates()
print("duplicates",dupe_df)
replace_df=merge_df.copy()
replace_df['maths']=replace_df['math_score'].replace({80:76,70:90})
print("replace",replace_df)
```

```python
import pandas as pd

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'score': [95, 67, 78, 88, 59],
    'grade': ['A', 'C', 'B', 'B', 'D']
})
df.index=['s1','s2','s3','s4','s5']
df.rename(columns={'name':'student','score':'marks'},inplace=True)
df['performance']=pd.cut(df['marks'],bins=[0,8,9,10,13],labels=['good','average','execellence','outsanding'],right=False)
sampling=df.sample(frac=1)
sample=df.sample(n=2)
df=pd.concat([df,pd.get_dummies(df['grade'],prefix='grade')],axis=1)
print(df)
print(sampling)
print(sample)
```
